# Outcome Classification Flow -- Sankey Diagram

Visualizes the confusion matrix as a flow diagram: GT outcomes (left) flowing
to algorithm outcomes (right). Correct classifications flow in the outcome's
color; misclassifications flow in red with counts labeled.

```python
from pathlib import Path
import json

snapshot_dir = Path(SNAPSHOT_DIR)
with open(snapshot_dir / 'metrics' / 'scalars.json') as f:
    scalars = json.load(f)

cm = scalars['outcome_label']['confusion_matrix']
print(f'Confusion matrix entries: {len(cm)}')
for k, v in sorted(cm.items(), key=lambda x: -x[1]):
    print(f'  {k}: {v}')
```
