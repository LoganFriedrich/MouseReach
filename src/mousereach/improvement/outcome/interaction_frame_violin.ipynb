# Interaction Frame Delta -- Violin Plot

Horizontal violin showing signed `interaction_frame_delta` (algo - GT)
for all touched-outcome segments where both sides committed.
Single distribution, jittered points overlaid, zero vertical reference line.

**Run via `_run_notebooks.py`** (this env lacks papermill).

```python
# ---- params ----
SNAPSHOT_DIR = r"Y:\2_Connectome\Behavior\MouseReach_Pipeline\Improvement_Snapshots\outcome\outcome_v2.4.4_pre_new_dlc"
SAVE = False
```

```python
from pathlib import Path
import pandas as pd
import numpy as np

snapshot_dir = Path(SNAPSHOT_DIR)
df = pd.read_csv(snapshot_dir / 'metrics' / 'outcome_per_segment.csv')

touched = df[
    df['gt_outcome'].isin(['retrieved', 'displaced_sa', 'displaced_outside']) &
    df['algo_outcome'].isin(['retrieved', 'displaced_sa', 'displaced_outside']) &
    df['interaction_frame_delta'].notna()
].copy()

deltas = touched['interaction_frame_delta'].astype(int).values
print(f'N eligible: {len(deltas)}')
print(f'Median |delta|: {int(np.median(np.abs(deltas)))}')
print(f'Mean signed: {np.mean(deltas):.1f}')
```
